Because Github nests data we need to manipulate our data to get actual fields for labels and milestones

```python
def get_label(row):
    labels = []
    if len(row) > 0:
        for r in row:
            labels.append(r['name'])
    else:
        labels.append('none')
    return labels
issues_df.loc[:, 'label_names'] = issues_df.labels.apply(get_label).reset_index(drop=True)
```

```python
alt.Chart(df_labels).mark_bar(size=1).encode(
    x=alt.X('created_at:T', axis=alt.Axis(title='Date Issue Opened')),
    y=alt.Y('issue_delta:Q', axis=alt.Axis(title='Time Issue Left Open in Days')),
    color=alt.Color('still_open', legend=alt.Legend(title="Issue Closed?"))
).properties(
    title="How Long is an Issue Open on PH Jekyll Repository"
).configure_axisY(
    titleAngle=0, 
    titleY=-10,
    titleX=-20,
    labelPadding=30, 
    labelAlign='left'
)
# chart.save('chart.png', scale_factor=2.0)
# save(chart, 'chart.png', method='selenium', webdriver=driver, scale_factor=2.0)
```
